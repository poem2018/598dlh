```python
# f = open('./proj_output100.txt', 'r') 
f = open('./output/mimic_diag_self_hi_nohy.out', 'r')   #
txt_content = f.readlines()
# print(txt_content)
# print(changeline_num)
# print(type(txt_content))
i = 0
f1_score = []
recall_10 = []
recall_20 = []
for i in range(len(txt_content)):
    str_ = txt_content[i]
    # print(str)
    if 'f1_score' in str_:
        temp = str_.split('f1_score')[1]
        f1_score.append(float(temp.split("top_k_recall")[0][1:-1]))
        list_ = temp.split("top_k_recall")[1][3:].split(' ')
        print(list_)
        recall_10.append(float(list_[0]))
        recall_20.append(float(list_[1]))

    i += 1

# print(f1_score)
print(recall_20)

f.close()
```

```python
from matplotlib import pyplot as plt
import math

fig = plt.figure(figsize=(12,5))
fig.suptitle("output results after trainning 92 epochs")
ax1 = fig.add_subplot()


x = np.linspace(0,len(f1_score),len(f1_score))
ax1.plot(x,f1_score, label = 'f1_score')
ax1.plot(x,recall_10, label = 'recall_10')
ax1.plot(x,recall_20, label = 'recall_20')

ax1.set_ylabel("f1_score, recall_10,recall_20")
ax1.legend(loc = 'upper left')

plt.savefig("./output_pic/mimic_dio_self_hi_nohy_recall.png")
plt.show()


print("max  recall_10",max(recall_10))
print("max  recall_20",max(recall_20))
print("max f1-score score",max(f1_score))
```

```python
# f = open('./proj_output100.txt', 'r') 
f = open('./output/mimic_hf_self_hi_hy.log', 'r')   #
txt_content = f.readlines()
# print(txt_content)
# print(changeline_num)
# print(type(txt_content))
i = 0
f1_score = []
auc = []
for i in range(len(txt_content)):
    str_ = txt_content[i]
    # print(str)
    if 'auc' in str_:
        temp = str_.split('auc')[1]
        f1_score.append(float(temp.split("f1_score")[1][1:-1]))
        auc.append(float(temp.split("f1_score")[0][1:-1]))

    i += 1

print(auc)

f.close()
```

```python
from matplotlib import pyplot as plt
import math

fig = plt.figure(figsize=(12,5))
fig.suptitle("output results after trainning 92 epochs")
ax1 = fig.add_subplot()


x = np.linspace(0,len(f1_score),len(f1_score))
ax1.plot(x,auc, label = 'auc')
ax1.plot(x,f1_score, label = 'f1_score')


ax1.set_ylabel("auc, f1_score")
ax1.legend(loc = 'upper left')

# plt.savefig("./output_pic/mimic_hf_self_hi_hy_02.png")
plt.show()


print("max auc score",max(auc))
print("max f1-score score",max(f1_score))
```
